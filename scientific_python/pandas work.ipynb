```python
gb = monthly.groupby(['station', 'year'])
gb
```

```python
len(gb)
```

```python
gb['precip']
```

```python
gb['precip'].sum()
```

```python
def stats(data):
    return pd.Series(
    [data.mean(), data.std()],
    index=['mean', 'stddev'],
    name=data.name)

gb['precip'].apply(stats)
```

```python
gb['precip'].apply(stats).unstack()
```
